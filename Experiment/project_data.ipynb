```python
x = df_by_env_sort_url_http[df_by_env_sort_url_http['httpVersion'] == 'HTTP/1.1'].reset_index(drop=True)['url'].values
x
```

```python
_y = {
    'onContentLoad': df_by_env['onContentLoad'].values,
    'onLoad': df_by_env['onLoad'].values,
    'cpu': df_by_env['cpu'].values,
    'ram': df_by_env['ram'].values,
}

df_by_version = {
    'http_1': df_by_env_sort_url_http[df_by_env_sort_url_http['httpVersion'] == 'HTTP/1.1'].reset_index(drop=True),
    'http_2': df_by_env_sort_url_http[df_by_env_sort_url_http['httpVersion'] == 'HTTP/2.0'].reset_index(drop=True),
    'http_3': df_by_env_sort_url_http[df_by_env_sort_url_http['httpVersion'] == 'HTTP/3'].reset_index(drop=True),
}

df_by_version['http_1']
#df_by_version['http_1']['cpu'].values
```

```python
fig = go.Figure()

# cpu, ram, onLoad, onContentLoad
y_axis = 'cpu'

fig.add_trace(go.Box(
    y=df_by_version['http_1']['cpu'].values,
    x=x,
    name='HTTP/1.1',
    marker_color='#3D9970'
))
fig.add_trace(go.Box(
    y=df_by_version['http_2']['cpu'].values,
    x=x,
    name='HTTP/2.0',
    marker_color='#FF4136'
))
fig.add_trace(go.Box(
     y=df_by_version['http_3']['cpu'].values,
    x=x,
    name='HTTP/3',
    marker_color='#FF851B'
))

fig.update_layout(
    yaxis_title=f'{y_axis}',
    boxmode='group' # group together boxes of the different traces for each value of x
)
fig.show()
```

#box graph (release)

```python
# 1, 2, 3
env_num = 3

# cpu, ram, onLoad, onContentLoad
y_axis = 'onContentLoad'

#layout_yaxis_range = None
layout_yaxis_range = [40,900]


df_by_env = df[df['env'] == env_num]
df_by_env_sort_url_http = df_by_env.sort_values(by=['url','httpVersion'], ascending=[True, True]).reset_index(drop=True)

x = df_by_env_sort_url_http[df_by_env_sort_url_http['httpVersion'] == 'HTTP/1.1'].reset_index(drop=True)['url'].values

df_by_version = {
    'http_1': df_by_env_sort_url_http[df_by_env_sort_url_http['httpVersion'] == 'HTTP/1.1'].reset_index(drop=True),
    'http_2': df_by_env_sort_url_http[df_by_env_sort_url_http['httpVersion'] == 'HTTP/2.0'].reset_index(drop=True),
    'http_3': df_by_env_sort_url_http[df_by_env_sort_url_http['httpVersion'] == 'HTTP/3'].reset_index(drop=True),
}

fig = go.Figure(layout_yaxis_range=layout_yaxis_range)
fig.add_trace(go.Box(
    y=df_by_version['http_1'][y_axis].values,
    x=x,
    name='HTTP/1.1',
    marker_color='#3D9970'
))
fig.add_trace(go.Box(
    y=df_by_version['http_2'][y_axis].values,
    x=x,
    name='HTTP/2.0',
    marker_color='#FF4136'
))
fig.add_trace(go.Box(
    y=df_by_version['http_3'][y_axis].values,
    x=x,
    name='HTTP/3',
    marker_color='#FF851B'
))

fig.update_layout(title=f'Env{env_num}',
    xaxis_title='site',
    yaxis_title=f'{y_axis}',
    boxmode='group'# group together boxes of the different traces for each value of x
)

fig.show()
```

#overlapping onLoad and onContentLoad (release)

```python
# site_1, site_2, site_3, site_4, site_5
site = 'site_5'

grouped_by_env_http = df[(df['url']==site)][['onContentLoad',  'onLoad','httpVersion', 'env',]].groupby(['httpVersion', 'env']).mean()
# grouped_by_env_http

bar_width = 0.2  # Adjust as needed

# env
x = ['1', '2', '3']

fig = go.Figure()
fig.add_trace(go.Bar(
    x=x,
    name='HTTP/1.1 onLoad',
    y=grouped_by_env_http.iloc[0:3]['onLoad'].values,
    offset=-bar_width-bar_width/2,
    width=bar_width,
    marker_color='#FF4136'
))
fig.add_trace(go.Bar(
    x=x,
    name='HTTP/1.1 onContentLoad',
    y=grouped_by_env_http.iloc[0:3]['onContentLoad'].values,
    offset=-bar_width-bar_width/2,
    width=bar_width,
    marker_color='#FF851B'
))

fig.add_trace(go.Bar(
    x=x,
    name='HTTP/2.0 onLoad',
    y=grouped_by_env_http.iloc[3:6]['onLoad'].values,
    offset=-bar_width/2,
    width=bar_width,
    marker_color='#04724D'
))
fig.add_trace(go.Bar(
    x=x,
    name='HTTP/2.0 onContentLoad',
    y=grouped_by_env_http.iloc[3:6]['onContentLoad'].values,
    offset=-bar_width/2,
    width=bar_width,
    marker_color='#52D053'
))

fig.add_trace(go.Bar(
    x=x,
    name='HTTP/3 onLoad',
    y=grouped_by_env_http.iloc[6:9]['onLoad'].values,
    offset=bar_width-bar_width/2,
    width=bar_width,
    marker_color='#765898'
))
fig.add_trace(go.Bar(
    x=x,
    name='HTTP/3 onContentLoad',
    y=grouped_by_env_http.iloc[6:9]['onContentLoad'].values,
    offset=bar_width-bar_width/2,
    width=bar_width,
    marker_color='#D4C2FC'
))

# Update layout
fig.update_layout(title=f'{site}',
                  xaxis_title='env',
                  yaxis_title='onLoad & onCountLoad')

# Show plot
fig.show()
```
